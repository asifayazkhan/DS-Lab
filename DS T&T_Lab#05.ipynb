```python
df.drop(['state','area_code','phone_number'],axis=1,inplace=True)
df.head()
```

```python
intl_plan_map = {'yes':1, 'no':0}
df['intl_plan'] = df['intl_plan'].map(intl_plan_map)

df
```

```python
voice_plan_map = {'yes':1, 'no':0}
df['voice_mail_plan'] = df['voice_mail_plan'].map(voice_plan_map)
df.tail()
```

```python
y = df.iloc[: , -1].values
X = df.iloc[: , : -1].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.3,
random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN= KNeighborsClassifier(n_neighbors=3)
KNN= KNN.fit(X_train, y_train)
y_predict= KNN.predict(X_test)
```

```python
print('Accuracy',KNN.score(X_train,y_train))
```

```python
from sklearn import metrics
print("Accuracy",metrics.accuracy_score(y_predict,y_test))
```

```python
kk=[]
ef=[]
for i in range(1,21):
    
    KNN= KNeighborsClassifier(n_neighbors=i)
    KNN= KNN.fit(X_train, y_train)
    y_predict= KNN.predict(X_test)
    eff=KNN.score(X_train,y_train)
    kk.append(i)
    ef.append(eff)
```

```python
import matplotlib.pyplot as plt
plt.plot(kk,ef)
```
